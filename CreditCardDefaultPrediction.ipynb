```python
pay_amt_columns = ["LIMIT_BAL","PAY_AMT1", "PAY_AMT2", "PAY_AMT3", "PAY_AMT4", "PAY_AMT5", "PAY_AMT6"]
dataset[pay_amt_columns] = dataset[pay_amt_columns].apply(pd.to_numeric, errors='coerce')


dataset["sum_pay_amt"] = dataset[pay_amt_columns].sum(axis=1)
```

```python
dataset["sum_pay_amt"]
```

```python
bill_amt_columns = ["BILL_AMT1", "BILL_AMT2", "BILL_AMT3", "BILL_AMT4", "BILL_AMT5", "BILL_AMT6"]
dataset[bill_amt_columns] = dataset[bill_amt_columns].apply(pd.to_numeric, errors='coerce')

dataset["sum_bill_amt"] = dataset[bill_amt_columns].sum(axis=1)

for i in range(len(dataset["sum_bill_amt"])):
    if dataset["sum_bill_amt"][i] == 0:
        dataset["sum_bill_amt"][i] = 1
```

```python
dataset["sum_bill_amt"]
```

```python
for i in range(len(pay_amt_columns)):
    dataset["division"] =  dataset['sum_bill_amt'] / dataset['sum_pay_amt']
```

# 03 Models Based upon division and default credit card prediction (Classification)

```python
X = dataset[["division"]]
y = dataset[["default.payment.next.month"]]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=1)
X_train = pca.fit_transform(X_train)
X_test = pca.transform(X_test)
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
lda = LDA(n_components = 1)
X_train = lda.fit_transform(X_train, y_train)
X_test = lda.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier= KNeighborsClassifier(n_neighbors = 50)
classifier.fit(X_train, y_train)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
y_pred = classifier.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

```python
# since we know that our data is imbalanced, let's apply SMOTE to balance the dataset and see if it improves our accuracy.
X = defaulters.drop('default.payment.next.month', axis=1)
y = defaulters['default.payment.next.month']

from imblearn.over_sampling import SMOTE

smote = SMOTE()
x_smote, y_smote = smote.fit_resample(X,y)

X_train, X_test, y_train, y_test = train_test_split(x_smote, y_smote, test_size=0.15, random_state=42)


rf = RandomForestClassifier()




rf.fit(X_train, y_train)




rf_pred = rf.predict(X_test)



rf_accuracy = accuracy_score(y_test, rf_pred)



print("Random Forest Accuracy:", rf_accuracy)
```
